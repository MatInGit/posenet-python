```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
body_part = "leftWrist"

for i in range(149):
    if i == 0:
        with open('output/pose'+str(i)+'.json') as json_file:
            data = json.load(json_file)
            if data["pose_scores"] >= .1:
                df = pd.DataFrame(data[body_part], index=[i])
    else:
        with open('output/pose'+str(i)+'.json') as json_file:
            data = json.load(json_file)
            if data["pose_scores"] >= .1:
                
                df_temp = pd.DataFrame(data[body_part], index=[i])
                
                df = pd.concat([df,df_temp])
            
        

df.describe()
```

```python
fig = plt.figure(figsize=(8,3), dpi=150)
plt.plot(df['x'],label = "x")
plt.plot(df['y'],label = "y")
plt.plot(np.sqrt(df['y']**2+df['x']**2),label = "abs")
plt.ylabel("Magnitude")
plt.xlabel("timestep")
plt.legend()


```

```python
fig = plt.figure(figsize=(8,3), dpi=150)
plt.plot(df['score'])
```
